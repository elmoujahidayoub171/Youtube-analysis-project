# Analyse univarié des variables quantitatives "subscribers" :

```python
subs_millions = df['subscribers'] / 1_000_000
```

```python
import numpy as np
from scipy.stats import gaussian_kde
from scipy.signal import find_peaks
# Créer la figure avec la courbe de densité
plt.figure(figsize=(14, 7))
# Créer et tracer la courbe de densité (KDE - Kernel Density Estimation)
density = gaussian_kde(subs_millions.dropna())
x_range = np.linspace(0, subs_millions.max(), 500)
y_density = density(x_range)
plt.plot(x_range, y_density, 'r-', linewidth=2.5, label='Courbe de densité (KDE)')
plt.title("Distribution des abonnés YouTube - Courbe de Densité", fontsize=14, fontweight='bold')
plt.xlabel("Nombre d'abonnés (millions)", fontsize=11)
plt.ylabel("Densité de probabilité", fontsize=11)
plt.legend(loc='upper right', fontsize=10)
plt.grid(True, alpha=0.3)
plt.xlim(0, 150)  # Zoom pour mieux voir la distribution
plt.tight_layout()
plt.show()
```

##  INTERPRETATION DE courbe de densité

1/Analyse de la forme de la distribution

-Asymétrie : La distribution est fortement asymétrique à droite (asymétrie positive).

-Modalité : La distribution est unimodale, présentant un pic unique 

-Type de distribution : Elle suit la loi exponentiel où la probabilité diminue rapidement à mesure que le nombre d'abonnés augmente.

2/Observation de la concentration des valeurs

-Zone de forte densité : La zone la plus large et la plus haute se situe en dessous de 20 millions d'abonnés, confirmant une concentration massive dans cette plage.

-Zone de faible fréquence : Au-delà de 40-50 millions, la courbe devient très fine  d'une très faible fréquence des comptes possédant un très grand nombre d'abonnés.

3/Évaluer visuellement la dispersion

 -un étalement important vers la droite indique une variabilité élevée.
