```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

# variable qualitative _ quantitative ( country _ video_views)

```python
df = pd.read_csv("/home/umoujahid/visualisation_project/data/Global-YouTube-Statistics_cleaned.csv")
```

```python
df["video views"].describe()
```

```python
#Top 3 countries 
top_3_countries = df["Country"].value_counts().head(3).index
top_3_countries
```

```python
#
filtered_df = df[df["Country"].isin(top_3_countries)]
```

```python
top3 = df["Country"].value_counts().head(3)
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=filtered_df, x="Country")
plt.title("Top 3 des pays les plus representes")
plt.xlabel("Pays")
plt.ylabel("Nombre de chaînes")
plt.show()
```

```python
plt.figure(figsize=(12, 6))

mean_views = filtered_df.groupby("Country", as_index=False)["video views"].mean()
sns.barplot(data=mean_views, x="Country", y="video views",palette="Set2")

plt.title("Vues moyennes par pays")
plt.xlabel("Pays")
plt.ylabel("Vues moyennes")
plt.show()
```

# Justification
Nous avons choisi de conserver uniquement les 3 pays les plus représentés dans le dataset.
On a retenu uniquement les top 3 countries afin de rendre l’analyse plus lisible et plus pertinente.

```python
#boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x="Country",y="video views",data=filtered_df)
plt.title("boxplot Country vs vido_views")
plt.show()
```

# Interpretation
Le boxplot met en évidence une forte variabilité du nombre de vues selon les pays étudiés. Bien que les médianes soient relativement proches, l’Inde et les États-Unis se distinguent par la présence de nombreuses valeurs aberrantes, traduisant l’existence de vidéos ayant généré un volume de vues exceptionnellement élevé. À l’inverse, la catégorie “Unknown” présente une dispersion plus faible et des valeurs extrêmes moins marquées

```python
#violin plot
plt.figure(figsize=(12, 6))
sns.violinplot(x="Country",y="video views",data=filtered_df)
plt.title("violinplot Country vs vido_views")
plt.show()
```

```python
#swarm plot
plt.figure(figsize=(12, 6))
sns.swarmplot(x="Country",y="video views",data=filtered_df)
plt.title("violinplot Country vs vido_views")
plt.show()
```

```python
#bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x="Country",y="video views",data=filtered_df)
plt.title("violinplot Country vs vido_views")
plt.show()
```

# Decision
l’Inde se distingue comme le pays de référence en termes de performance des vues. Il serait alors pertinent d’identifier les facteurs à l’origine de ces performances et de s’en inspirer pour améliorer et optimiser le contenu proposé dans les autres pays, dont les résultats restent relativement plus faibles.
